```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
df = pd.read_csv('/content/ShartankIndiaAllPitches.csv')

# Line Plot
plt.plot(df['Episode Number'], df['Investment Amount (In Lakhs INR)'])
plt.xlabel('Episode Number')
plt.ylabel('Investment Amount (In Lakhs INR)')
plt.title('Investment Amount by Episode Number')
plt.show()

# Scatter Plot
plt.scatter(df['Debt (In lakhs INR)'], df['Equity'])
plt.xlabel('Debt (In lakhs INR)')
plt.ylabel('Equity')
plt.title('Debt vs Equity')
plt.show()

# Bar Plot
plt.bar(df['Brand'], df['Investment Amount (In Lakhs INR)'])
plt.xlabel('Brand')
plt.ylabel('Investment Amount (In Lakhs INR)')
plt.title('Investment Amount by Brand')
plt.xticks(rotation=90, fontsize=4)
plt.show()
```

```python
sns.lineplot(x='Episode Number', y='Investment Amount (In Lakhs INR)', data=df)
plt.show()
```

```python
sns.scatterplot(x='Equity', y='Investment Amount (In Lakhs INR)', data=df)
plt.show()
```

```python
columns = ['Episode Number', 'Brand', 'Investment Amount (In Lakhs INR)', 'Debt (In lakhs INR)', 'Equity',
           'Anupam', 'Ashneer', 'Namita', 'Aman', 'Peyush', 'Vineeta', 'Ghazal', 'Season']
heatmap_data = df[columns]
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
equity = df['Equity'].str.rstrip("%").astype(float)/100
```

```python
equity_per_episode = (equity.groupby(df['Episode Number']).mean() * 100).astype(int)
equity_average = (equity.mean() * 100).astype(int)
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x = equity_per_episode.index, y = equity_per_episode.values, palette="Paired" )
plt.ylabel('Average Equity holding in %')
```

On an **average, 9%** of **shares** were **bought by sharks in the whole season.** On **episode basis**, there is seen too much variation. **highest share** were **brought** **in episode 21** which is **37%**.

```python
invest = df['Investment Amount (In Lakhs INR)'].sum()
```

```python
debt = df['Debt (In lakhs INR)'].sum()
```

```python
import plotly.graph_objs as go
```

```python
groups = ['Investment Amount (In Lakhs INR)','Debt (In lakhs INR)']
values = [invest, debt]
values
```

```python
trace = go.Pie(labels = groups, values=values,textfont=dict(size=15),marker=dict(line=dict(color='#000000', width = 4)))
iplot([trace])
```

Season 1 saw a huge investment in terms of equity for share. Almost **2537 lakh rupees** was invested by sharks which makes **89.4% of all the money** given to pitchers. **Debt holds** a small share for investment yet, it has significant **share of 10.6% or 301 lakh rupees**

```python
avg_invest = df['Investment Amount (In Lakhs INR)'].groupby(df['Episode Number']).mean().astype(int)
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x=avg_invest.index, y=avg_invest.values)
plt.ylabel('Average Investment per episode in Lakhs')
plt.show()
```

In terms of Investment, **Episode 1** was **highly successful** with average **investment of 46 lakh**. In **episode 18**, sharks didn't made any investment which is **worst day of the season**
